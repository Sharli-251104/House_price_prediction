```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("Housing.csv")
data
```

```python
columns_to_encode = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']
```

```python
label_encoders = {}
for column in columns_to_encode:
    label_encoders[column] = LabelEncoder()
    data[column] = label_encoders[column].fit_transform(data[column])
```

```python
x = data.drop(columns=['price'])
```

```python
x
```

```python
x.columns = ['feature_' + str(i) for i in range(len(x.columns))]
```

```python
x
```

```python
y = data['price']
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
data_regressor = DecisionTreeRegressor(random_state= 42)
```

```python
data_regressor.fit(x_train, y_train)
```

```python
y_pred = data_regressor.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
print("Mean Squared Error:", mse)
```

```python
new_features = x_test.iloc[0].values.reshape(1,-1)
```

```python
x_test.iloc[0]
```

```python
predicted_price = data_regressor.predict(new_features)
```

```python
print("Predicted price for new house:", predicted_price[0])
```
